```python
from mpi4py import MPI
import ray_py
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n = 10000
r = np.random.random((n,3))-0.5
r[:,2]*=0.01
h = np.ones(n)*1.e-2
op = np.ones(n)
```

```python
tau=ray_py.calc_depths(r,h,op)
```

```python
plt.scatter(r[:,0],r[:,1],c=tau,s=0.1)
plt.colorbar()
```
